```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def correlacion_pearson(data_frame, columna1, columna2):
    return np.corrcoef(data_frame[columna1], data_frame[columna2])[0,1];
```

# Datasets
Para los experimentos se crean las siguientes familias de instancias para evaluar los distintos métodos.
1. **todos-los-productos-agregables**: Son instancias donde se pueden agregar todos los productos.
2. **ningun-producto-agregable**: Son instancias donde no se puede agregar ningun producto.
3. **R-grande**: Instancias con un R muy alto.
4. **random**: Instancias con valores aleatorios.
5. **random-muchos-productod**: Instancias con valores aleatorios pero mas productos. Hasta 141

# Métodos
Los métodos evaluados en este notebook son los siguientes:
- **FB**: Fuerza Bruta.
- **BTPodas**: Backtracking con podas por optimalidad y factibilidad.
- **PD**: Programación dinámica.

# Experimentos
Leemos el dataframe con todos los resultados.

```python
df_resultados = pd.read_csv("resultado.csv");
df_resultados.head()
```

## Experimento 1: Complejidad y analisis de Fuerza Bruta
En este experimento evaluamos el método FB con los datasets todos-los-productos-agregables y ningun-producto-agregable

```python
# Graficamos los tiempos de ejecución de cada dataset en función de n.
df_complejidad_FB = df_resultados[df_resultados["Experimento"]=="Maximo"].copy();
fig = sns.scatterplot(data=df_complejidad_FB, x='tM', y='tiempoM', hue='dataset');
fig.set(xlabel='threads', ylabel='tiempo (ms)');
plt.savefig("maximo-threads.svg");
```

```python
df_complejidad_FB = df_resultados[(df_resultados["metodo"]=="FB") & (df_resultados["dataset"]=="todos-los-productos-agregables")].copy();
df_complejidad_FB["tiempo_esperado"] = df_complejidad_FB["n"] * (2.0 ** df_complejidad_FB["n"]) * 0.0000015; # Exponencial
fig = sns.scatterplot(data=df_complejidad_FB, x='n', y='tiempo');
fig = sns.lineplot(data=df_complejidad_FB, x='n', y='tiempo_esperado', color="orange");
fig.set(yscale="log")
fig.legend(labels=['O(n*2^n)','FB']);
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/fb-complejidad-logaritmico.svg")
```

```python
df_complejidad_FB = df_resultados[(df_resultados["metodo"]=="FB") & (df_resultados["dataset"]=="todos-los-productos-agregables")].copy();
df_complejidad_FB["complejidad"] = df_complejidad_FB["n"] * 2.0 ** df_complejidad_FB["n"];
fig = sns.lmplot(data=df_complejidad_FB, x="tiempo", y="complejidad");
plt.savefig("graficos/fb-correlacion.svg");
print("Índice de correlación de Pearson:", correlacion_pearson(df_complejidad_FB, "tiempo", "complejidad"));
```

## Experimento 2: Complejidad de Backtracking
En este experimento evaluamos el método BT con los datasets mejor-caso-bt y peor-caso-bt. La hipótesis es que va a tener un comportamiento exponencial en el peor caso y lineal en el mejor caso.

```python
df_aleatorio = df_resultados[((df_resultados["metodo"]=="BTPodas") | (df_resultados["metodo"]=="PD")) & (df_resultados["dataset"] == "random")].copy();
fig = sns.scatterplot(data=df_aleatorio, x='n', y='tiempo', hue='metodo');
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/comparacion-bt-fb-dp.svg");
```

## Experimento 5: Comparación FB y Programación Dinámica
En este experimento se trata de comparar FB y PD cuando hay pocos productos, pero el R es muy alto

```python
df_R_grande = df_resultados[((df_resultados["metodo"]=="FB") | (df_resultados["metodo"]=="PD")) & (df_resultados["dataset"] == "R-grande")].copy();
fig = sns.scatterplot(data=df_R_grande, x='n', y='tiempo', hue='metodo');
fig.set(xlabel='n', ylabel='tiempo (ms)');
plt.savefig("graficos/R-grande-BT-PD.svg");
```
