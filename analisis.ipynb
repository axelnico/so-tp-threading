# Datasets
Para los experimentos se crean las siguientes familias de instancias para evaluar los distintos métodos.
1. **todoAleatorio**: Todas palabras aleatorias.
2. **todosIguales**: La misma palabra.
3. **letrasInicialUniforme**: Distribucion informe de primera letra en palabras.
4. **primeraLetraIgual**: Palabras que empiezan con l.

# Métodos
Los métodos evaluados en este notebook son los siguientes:
- **Maximo**: Maximo concurrente.
- **Lectura**: Carga de palabras de archivos

# Experimentos
Leemos el dataframe con todos los resultados.

```python
df_resultados = pd.read_csv("resultado.csv");
df_resultados.head()
```

## Experimentos: Analisis de maximo
En estos experimentos evaluamos el tiempo de ejecucion de la funcion maximo concurrente en funcion de la cantidad de threads

```python
# Graficamos los tiempos de ejecución de cada dataset en función de threads.
df_complejidad_FB = df_resultados[(df_resultados["Experimento"]=="Maximo") & (df_resultados["archivos"]=="['todosIguales.txt']")].copy();
fig = sns.scatterplot(data=df_complejidad_FB, x='tM', y='tiempoM', hue='archivos');
fig.set(xlabel='threads', ylabel='tiempo (ms)');
plt.savefig("maximo-threads-muchas-palabras-todos-iguales.svg");
```

```python
df_complejidad_FB = df_resultados[(df_resultados["Experimento"]=="Maximo") & (df_resultados["archivos"]=="['todoAleatorio.txt']")].copy();
fig = sns.scatterplot(data=df_complejidad_FB, x='tM', y='tiempoM', hue='archivos');
fig.set(xlabel='threads', ylabel='tiempo (ms)');
plt.savefig("maximo-threads-muchas-palabras-todo-aleatorio.svg");
```

```python
df_complejidad_FB = df_resultados[(df_resultados["Experimento"]=="Maximo") & (df_resultados["archivos"]=="['primeraLetraIgual.txt']")].copy();
fig = sns.scatterplot(data=df_complejidad_FB, x='tM', y='tiempoM', hue='archivos');
fig.set(xlabel='threads', ylabel='tiempo (ms)');
plt.savefig("maximo-threads-muchas-palabras-primera-letra-igual.svg");
```

```python
df_complejidad_FB = df_resultados[(df_resultados["Experimento"]=="Maximo") & (df_resultados["archivos"]=="['letrasInicialUniforme.txt']")].copy();
fig = sns.scatterplot(data=df_complejidad_FB, x='tM', y='tiempoM', hue='archivos');
fig.set(xlabel='threads', ylabel='tiempo (ms)');
plt.savefig("maximo-threads-muchas-palabras-letras-inicial-uniforme.svg");
```
